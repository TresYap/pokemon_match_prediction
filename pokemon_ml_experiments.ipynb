```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
train.head(1)
```

```python
#Drop unnecessary columns
try:
#     drop = ["first_Type 1", "first_Type 2", "second_Type 1", "second_Type 2", "First_pokemon", "Second_pokemon", "first_Multiplier", "second_Multiplier"]
    drop = ["First_pokemon", "Second_pokemon"]
    train.drop(drop, axis=1, inplace=True)
except Exception as e:
    print(e)
```

```python
train
```

```python
y = train.outcome
train.drop("outcome", axis=1, inplace=True)
```

```python
reg_model = LogisticRegression()
svm_model = LinearSVC(max_iter=10000)
rf_model = RandomForestClassifier(n_estimators=300)
gb_model = GradientBoostingClassifier(n_estimators=150)

reg_scores = cross_val_score(reg_model, train, y, cv=5, scoring="f1")
svm_scores = cross_val_score(svm_model, train, y, cv=5, scoring="f1")
rf_scores = cross_val_score(rf_model, train, y, cv=5, scoring="f1")
gb_scores = cross_val_score(gb_model, train, y, cv=5, scoring="f1")
```

```python
print("Cross-Validation F1 Scores (mean):")
print("Logistic Regression: %0.2f" % reg_scores.mean())
print("SVM: %0.2f" % svm_scores.mean())
print("Random Forest: %0.2f" % rf_scores.mean())
print("Gradient Boosting: %0.2f" % gb_scores.mean())
```
